### 3. Low Level Data Understanding : 
#### a. Find count of unique values in location column. 
####b. Find which continent has maximum frequency using values counts. 
####c. Find maximum & mean value in 'total_cases'. 
####d. Find 25%,50% & 75% quartile value in 'total_deaths'. 
####e. Find which continent has maximum 'human_development_index'. 
####f. Find which continent has minimum 'gdp_per_capita'. 

```python
# a. Find count of unique values in location column.
df.location.nunique()
```

```python
# b. Find which continent has maximum frequency using values counts.
b= df['continent'].dropna()
x= b.value_counts()
for i in b:
  if x[i]==x.max():
    print(f"{i} continent has maximum frequency :",x.max())
    break
```

```python
# c. Find maximum & mean value in 'total_cases'.

print("maximum in total_cases :",df.total_cases.max())
print("mean in total_cases :",df.total_cases.mean())
```

```python
# d. Find 25%,50% & 75% quartile value in 'total_deaths'.

df['total_deaths'].quantile([0.25,0.50,0.75])
```

```python
# e. Find which continent has maximum 'human_development_index'.

df_1=df[['continent','human_development_index']].dropna()
df_1[df_1['human_development_index']==df_1['human_development_index'].max()].drop_duplicates()
```

```python
# f. Find which continent has minimum 'gdp_per_capita'.
df_1=df[['continent','gdp_per_capita']].dropna()
df_1[df_1['gdp_per_capita']==df_1['gdp_per_capita'].min()].drop_duplicates()
```

###4. Filter the dataframe with only this columns 
['continent','location','date','total_cases','total_deaths' 'gdp_per_capita',' human_development_index'] 

and update the data frame.

```python
df_1=df.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
df_1
```

### 5. Data Cleaning 
a. Remove all duplicates observations 

b. Find missing values in all columns 

c. Remove all observations where continent column value is missing 

Tip : using subset parameter in dropna 

d. Fill all missing values with 0

```python
# a. Remove all duplicates observations 
df_1.drop_duplicates()
```

```python
# b. Find missing values in all columns
df_1.isnull()
```

```python
# c. Remove all observations where continent column value is missing

df_2=df_1.dropna(subset=['continent'])
```

```python
df_2
```

```python
# d. Fill all missing values with 0

df_3=df_2.fillna(0)
```

```python
df_3
```

### 7. Data Aggregation: 
a. Find max value in all columns using groupby function on 'continent' column Tip: use reset_index() after applying groupby

b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)

```python
# a. Find max value in all columns using groupby function on 'continent' column Tip: use reset_index() after applying groupby

df_3.reset_index(inplace = True)
x=df_3.groupby(df_3['continent']).apply(max)
x
```

```python
# b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)

df_groupby =x
df_groupby
```

### 8. Feature Engineering : 
a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases']=df_groupby["total_deaths"]/df_groupby['total_cases']
```
